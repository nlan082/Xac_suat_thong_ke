```python
import pandas as pd
df = pd.read_excel("HISTOGRAM.xlsx")
df.head()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Giả sử df là DataFrame chứa dữ liệu đã đọc từ file Excel
# df['gioi_tinh'] = 0 (nam), 1 (nữ)
# df['chi_tieu'] = 0–4

nam = df[df['gioi_tinh'] == 0]['chi_tieu']
nu = df[df['gioi_tinh'] == 1]['chi_tieu']

# Kiểm định t
t_stat, p_value = ttest_ind(nam, nu, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Có sự khác biệt có ý nghĩa thống kê giữa nam và nữ.")
else:
    print("Không có sự khác biệt đáng kể giữa nam và nữ.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file Excel
df = pd.read_excel("HISTOGRAM.xlsx")

# Vẽ histogram với dữ liệu đã mã hoá
plt.figure(figsize=(8, 5))
plt.hist(df["chi_tieu_thang_mahoa"], bins=[-0.5, 0.5, 1.5, 2.5, 3.5, 4.5], edgecolor='black', color='lightcoral', rwidth=0.9)

plt.title("Biểu đồ Histogram mức chi tiêu mỹ phẩm (mã hóa)")
plt.xlabel("Mức chi tiêu (mã hóa: 0–4)")
plt.ylabel("Số người")
plt.xticks([0, 1, 2, 3, 4])
plt.grid(axis='y')
plt.show()
```
