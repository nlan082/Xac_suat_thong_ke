```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the table
data = {
    'Loại mỹ phẩm sử dụng': ['Trang điểm', 'Chăm sóc da', 'Chăm sóc tóc', 'Chăm sóc cơ thể'],
    'Số người': [100, 171, 136, 103],
    'Tỷ lệ (%)': [50.51, 86.36, 68.69, 52.02]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for the number of people
bars = ax.bar(df['Loại mỹ phẩm sử dụng'], df['Số người'], color='skyblue')

# Annotate with the percentage values on the right side
for bar, percentage in zip(bars, df['Tỷ lệ (%)']):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval, f'{percentage}%', ha='left', va='bottom', fontsize=12)

# Labels and title
ax.set_xlabel('Loại mỹ phẩm sử dụng')
ax.set_ylabel('Số người')
ax.set_title('Số người sử dụng các loại mỹ phẩm')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Đọc file Excel (upload file trước vào Colab nếu chưa có)
df = pd.read_excel("/content/xstk_dulieumahoa.xlsx")

# Xem trước dữ liệu
df.head()

cols_for_clustering = [
    'thu_nhap_thang_mahoa',
    'chi_tieu_thang_mahoa',
    'thuong_dung_my_pham_mahoa',
    'cham_soc_da',
    'cham_soc_toc',
    'cham_soc_co_the'
]

X = df[cols_for_clustering]

from sklearn.cluster import KMeans

# Chọn số nhóm K
kmeans = KMeans(n_clusters=3, random_state=42)
df['nhom_phan_loai'] = kmeans.fit_predict(X)

# Xem kết quả
df[['nhom_phan_loai'] + cols_for_clustering].head()

df['nhom_phan_loai'].value_counts()

df.groupby('nhom_phan_loai')[cols_for_clustering].mean()

df.to_excel("du_lieu_da_phan_nhom.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt

# Đếm số lượng người mỗi nhóm
counts = df['nhom_phan_loai'].value_counts().sort_index()

# Vẽ biểu đồ cột
fig, ax = plt.subplots(figsize=(8,6))
bars = ax.bar(counts.index.astype(str), counts.values, color='skyblue')

# Thêm tiêu đề và nhãn
ax.set_title('Số lượng người theo từng nhóm phân loại (K-means)', fontsize=14)
ax.set_xlabel('Nhóm phân loại', fontsize=12)
ax.set_ylabel('Số lượng người', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Tạo chú thích bên phải
group_labels = {
    '0': 'Nhóm 0: Quan tâm đến làm đẹp nhiều',
    '1': 'Nhóm 1: Quan tâm đến làm đẹp trung bình',
    '2': 'Nhóm 2: Quan tâm đến làm đẹp ít'
}

# Chú thích ứng với số nhóm bạn đang có
handles = [plt.Rectangle((0,0),1,1, color='skyblue')] * len(counts)
labels = [group_labels[str(i)] for i in counts.index]

# Thêm legend bên phải
ax.legend(handles, labels, title='Chú thích nhóm', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

group_means = df.groupby('nhom_phan_loai')[cols_for_clustering].mean().reset_index()
group_means = pd.melt(group_means, id_vars='nhom_phan_loai')

plt.figure(figsize=(10,6))
ax = sns.barplot(x='value', y='variable', hue='nhom_phan_loai', data=group_means)
plt.title('So sánh đặc điểm giữa các nhóm phân loại')
plt.xlabel('Giá trị trung bình')
plt.ylabel('Biến')
plt.grid(axis='x')

# Chỉnh legend sang bên phải
plt.legend(title='Nhóm phân loại', loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
